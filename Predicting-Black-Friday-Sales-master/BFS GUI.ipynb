# Black Friday Sales GUI

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
traindf = pd.read_csv(r"traindf.csv")
testdf = pd.read_csv(r"testdf.csv")
```

```python
train1 = pd.read_pickle('traindf_clean.pkl')
test1 = pd.read_pickle('testdf_clean.pkl')
```

```python
 def citystay(stay):
                if stay == '0':
                     return 0
                elif stay == '1':
                    return 1
                elif stay == '2':
                    return 2
                elif stay == '3':
                    return 3
                else:
                    return 4
```

```python
from IPython.display import display
import plotly.express as px
from ipywidgets import widgets


#visualizations
print("BLACK FRIDAY SALES")
print("\n")
print("Visualizations-")
print("Enter your option in the textbox:")
print("a. Correlation with respect to 'Purchase'")
print("b. Purchase amount distribution")
print("c. Purchase number in black friday based on Gender")
print("d. Purchase number in black friday based on Age")
print("e. Purchase number in black friday based on Occupation")
print("f. Purchase number in black friday based on City Category")
print("g. Purchase number in black friday based on Stay in Current City years")
print("h. Purchase number in black friday based on Marital Status")

text = widgets.Text()
display(text)  

def handle_submit(sender):
    print(text.value)
    if text.value == "a":
        corrIndex = traindf.corr().nlargest(10,'Purchase')['Purchase'].index
        corr = traindf[corrIndex].corr()
        plt.figure(figsize=(16,8))
        ax = sns.heatmap(corr,annot=True,cmap="YlGnBu")
        bottom, top = ax.get_ylim()
        ax.set_ylim(bottom + 0.5, top - 0.5)
        plt.show()
    elif text.value == "b":
        plt.style.use(style = 'fivethirtyeight')
        plt.figure(figsize=(10,5))
        sns.distplot(traindf['Purchase'], bins = 30)
        plt.xlabel('Amount spent in Purchase')
        plt.ylabel('Number of Customers')
        plt.title('Purchase amount Distribution')
    elif text.value == "c":
        malesPurchaserData = traindf.loc[traindf['Gender'] == 'M']
        femalesPurchaserData = traindf.loc[traindf['Gender'] == 'F']
        genderCountData = [malesPurchaserData.shape[0],femalesPurchaserData.shape[0]]
        labels=['Males','Females']
        plt.axis('equal')

        plt.pie(genderCountData, labels=labels,
                      explode=[0.1,0],
                      autopct='%1.1f%%',
                      shadow=True,
                      startangle=0,
                      labeldistance=1.1,
                      pctdistance=.6)

        plt.legend(labels)
        plt.title('Purchaser number in black friday based on Gender')
        plt.show()
    elif text.value == "d":
        sns.countplot(x='Age', data=traindf, palette='RdBu_r')
        
    elif text.value == "e":
        traindf[['Occupation']].hist(bins=20)
    elif text.value == "f":
        
        cityA = traindf.loc[traindf['City_Category'] == 'A']
        cityB = traindf.loc[traindf['City_Category'] == 'B']
        cityC = traindf.loc[traindf['City_Category'] == 'C']
        City = [cityA.shape[0],cityB.shape[0],cityC.shape[0]]
        labels=['cityA','cityB','cityC']
        plt.axis('equal')

        plt.pie(City, labels=labels,
                      explode=[0.1,0.05,0.05],
                      autopct='%1.1f%%',
                      shadow=True,
                      startangle=0,
                      labeldistance=1.1,
                      pctdistance=.6)

        plt.legend(labels)
        plt.title('Purchase in black friday based on City_Category')
        plt.show()
    
    elif text.value == "g":
        traindf['Stay_In_Current_City_Years'] = traindf['Stay_In_Current_City_Years'].apply(citystay)
        testdf['Stay_In_Current_City_Years'] = testdf['Stay_In_Current_City_Years'].apply(citystay)
        sns.distplot(traindf.Stay_In_Current_City_Years, kde=True);
        
                
                    
        
    elif text.value == "h":
            sns.countplot(x='Marital_Status', data=traindf, palette='RdBu_r')

            
        
    else:
        print("enter a valid option")
    
text.on_submit(handle_submit)    












X = train1.iloc[:, :-1]
Y = train1["Purchase"]

from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test= train_test_split(X, Y, test_size = 0.20, random_state = 0)

from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.fit_transform(X_test)

from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.preprocessing import PolynomialFeatures
from sklearn.ensemble import GradientBoostingRegressor


from ipywidgets import Button, Layout
from IPython.display import display

print("Click the button to see the accuracy of any model:")
print("(Might take some time to run)")

button = Button(description='Multiple Regression',
           layout=Layout(width='30%', height='40px'))
display(button)

button1 = Button(description='Polynomial Regression',
           layout=Layout(width='30%', height='40px'))
display(button1)

button2 = Button(description='Random Forest',
           layout=Layout(width='30%', height='40px'))
display(button2)

button3 = Button(description='Gradient Boosting',
           layout=Layout(width='30%', height='40px'))
display(button3)


def but(b):
    
    print("Multiple Regression")
    # Fitting Multiple Regression to Dataset
    Linearreg = LinearRegression()
    Linearreg.fit(X_train, Y_train)

    # Predicting the Purchase Value for Linear Regressor
    y_predMR = Linearreg.predict(X_test)

    # finding the mean_squared error
    mse = mean_squared_error(Y_test, y_predMR)
    print("RMSE Error:", np.sqrt(mse))

   
button.on_click(but)  
button.style.button_color = 'pink'

def but1(a):
    print("Polynomial Regression")
    
    # Fitting Polynomial Regression to Dataset
    poly_reg = PolynomialFeatures(degree = 3)
    X_poly = poly_reg.fit_transform(X_train)
    regressor = LinearRegression()
    regressor.fit(X_poly, Y_train)

    # Predicting the Purchase Values for Polynomial Regressor
    y_predPR = regressor.predict(poly_reg.fit_transform(X_test))


    # Root_mean_squared_error
    from math import sqrt
    rmse = sqrt(mean_squared_error(Y_test, y_predPR))
    print(rmse)
button1.on_click(but1)  
button1.style.button_color = 'pink'

def but2(c):
    print("Decision Tree")
    
    # Fitting Decision Tree Regression to Dataset
    Tree_regressor = DecisionTreeRegressor(random_state = 5)
    Tree_regressor.fit(X_train, Y_train)

    # Predicting Purchase value for Decision Tree Regressor
    y_predDTR = Tree_regressor.predict(X_test)

    # finding the mean_squared error
    mse = mean_squared_error(Y_test, y_predDTR)
    print("RMSE Error:", np.sqrt(mse))

button2.on_click(but2)  
button2.style.button_color = 'pink'

def but3(d):
    # Fitting Gradient Boosting Regression to Dataset
    model = GradientBoostingRegressor(n_estimators= 600,max_depth = 9)
    model.fit(X_train, Y_train)

    # Predicting Purchase value from Gradient Boosting Regressor
    y_predGBR = model.predict(X_test)


    # finding the mean_squared error
    mse = mean_squared_error(Y_test, y_predGBR)
    print("RMSE Error:", np.sqrt(mse))

button3.on_click(but3)  
button3.style.button_color = 'pink'

```
